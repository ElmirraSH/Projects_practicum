## Исследование использования сервиса Яндекс Афиша


## Цели и задачи проекта

**Цель:** 
 Проанализировать данные, чтобы предположить, почему продажи билетов на одни мероприятия неожиданно растут, а на другие снижаются. Сделать выводы об изменении пользовательских предпочтений. Проверить гипотезу о разнице в поведении пользователей с мобильных и стационарных устройств. 

## Описание данных

Предоставлены данные о бронировании билетов на сервисе Яндекс Афиша за период с 1 июня по 31 октября 2024 года. 

Датасеты:

`final_tickets_orders_df.csv` включает информацию обо всех заказах билетов, совершённых с двух типов устройств — мобильных и стационарных. Датасет содержит такие поля:
- order_id — уникальный идентификатор заказа.
- user_id — уникальный идентификатор пользователя.
- created_dt_msk — дата создания заказа (московское время).
- created_ts_msk — дата и время создания заказа (московское время).
- event_id — идентификатор мероприятия из таблицы events.
- cinema_circuit — сеть кинотеатров. Если не применимо, то здесь будет значение 'нет'.
- age_limit — возрастное ограничение мероприятия.
- currency_code — валюта оплаты, например rub для российских рублей.
- device_type_canonical — тип устройства, с которого был оформлен заказ, например mobile для мобильных устройств, desktop для стационарных.
- revenue — выручка от заказа.
- service_name — название билетного оператора.
- tickets_count — количество купленных билетов.
- total — общая сумма заказа.
- days_since_prev — количество дней с предыдущей покупки для каждого пользователя. Если покупки не было, то данные содержат пропуск.


`final_tickets_events_df.csv` содержит информацию о событиях, включая город и регион события, а также информацию о площадке проведения мероприятия. 
Датасет содержит такие поля:
- event_id — уникальный идентификатор мероприятия.
- event_name — название мероприятия. 
- event_type_description — описание типа мероприятия.
- event_type_main — основной тип мероприятия: театральная постановка, концерт и так далее.
- organizers — организаторы мероприятия.
- region_name — название региона.
- city_name — название города.
- venue_id — уникальный идентификатор площадки.
- venue_name — название площадки.
- venue_address — адрес площадки.

`final_tickets_tenge_df.csv` с информацией о курсе тенге к российскому рублю за 2024 год. Значения в рублях представлено для 100 тенге. Датасет содержит такие поля:
- nominal — номинал (100 тенге);
- data — дата;
- curs — курс тенге к рублю;
- cdx — обозначение валюты (kzt).
Эти данные необходимы для конвертации валюты.

## 2. Предобработка данных и подготовка их к исследованию

Объединим датасеты, используя left join. Каждая строка в этих датафреймах — это информация о заказе, поэтому такие данные удобно соединить в один датафрейм, который можно использовать для поиска закономерностей в данных. Соединять данные будем по идентификатору события, сохранив все заказы.

```python
df = df_orders.merge(df_events, on='event_id', how='left').merge(df_currency, left_on='created_dt_msk', right_on='data', how='left')
```

```python
df.head()
```

```python
df.info()
```

```python
#проверим значения категориальных данных. 
for column in ['age_limit','currency_code','device_type_canonical', 'service_name','event_type_description', 'event_type_main']:
    print(f'Уникальные значений в столбце {column}:')
    print(df[column].unique())
```

В столбцах `event_type_description`, `event_type_main` есть значения `nan`, которые могут обозначать пропуски в данных или отсутствие информации. 

```python
#подсчитаем количество значений `nan` в следующих столбцах
for column in ['service_name','event_type_description', 'event_type_main','region_name','city_name', 'venue_name'  ]:
    count_nan = df[column].isna().sum()
    print(f'Количество значений nan в {column}:')
    print(count_nan)
```

```python
df_nan = df[pd.isna(df['event_type_description'])]
print(df_nan)
```

```python
df['event_type_description'].isna().sum()/df.shape[0]*100
```

Количество строк с `nan` одинаковое по вышепроверенным столбцам (238), при выводе таблицы `nan` по столцам в одних и тех же строках. Относительное количество этих строк 0,08%, следовательно можем их удалить

```python
df_without_nan = df.dropna(subset= ['event_type_main'])
df_without_nan.shape[0]
```

Для анализа данных важными показателями являются выручка с заказа `revenue` и количество билетов в заказе `tickets_count`, поэтому проверим их распределение и наличие выбросов 

```python
df_without_nan['tickets_count'].describe()
```

```python
plt.figure(figsize=(10, 2))
df_without_nan['tickets_count'].plot(kind='box', vert=False)
plt.title('Распределение количества билетов в заказе')
plt.xlabel('Количество билетов')
plt.show()
```

```python
df_without_nan.loc[df_without_nan['currency_code']=='rub']['revenue'].describe()
```

Есть отрицательные значения и выбросы, отберем значения по 99-му процентилю

```python
df_without_nan.loc[df_without_nan['currency_code']=='rub']['revenue'].quantile(0.99)
```

```python
df_rub= df_without_nan.loc[df_without_nan['currency_code']=='rub']
ninety_ninth_percentile = df_rub['revenue'].quantile(0.99)

filtered_df = df_rub[df_rub['revenue'] <= ninety_ninth_percentile]
print(f'При отборе значений по 99 процентилю (валюта rub), удалено {df_rub.shape[0]-filtered_df.shape[0]} строк, что составляет {round((df_rub.shape[0]-filtered_df.shape[0])/df_rub.shape[0]*100,2)}%')
```

```python
df_without_nan.loc[df_without_nan['currency_code']=='kzt']['revenue'].describe()
```

```python
df_kzt= df_without_nan.loc[df_without_nan['currency_code']=='kzt']
ninety_ninth_percentile_kzt = df_kzt['revenue'].quantile(0.99)

filtered_df_kzt = df_kzt[df_kzt['revenue'] <= ninety_ninth_percentile_kzt]
print(f'При отборе значений по 99 процентилю (валюта kzt), удалено {df_kzt.shape[0]-filtered_df_kzt.shape[0]} строк, что составляет {round((df_kzt.shape[0]-filtered_df_kzt.shape[0])/df_kzt.shape[0]*100,2)}%')
```

```python
#объединим датафреймы, отобранные по валютам
df_all = pd.concat([filtered_df_kzt, filtered_df])
df_all = df_all.reset_index(drop=True)
```

```python
#Проверим наличие полных дубликатов
duplicates=df_all.duplicated().sum()
print(f"Полных дубликатов {duplicates}")
```

```python
duplicates_headen = df_all[df_all.columns.drop('order_id')].duplicated().sum()
duplicates_share=round(duplicates_headen/df_all.shape[0]*100,2)
print(f"Дубликатов по бронированию билета без учёта идентификаторов заказа {duplicates_headen}, что составляет {duplicates_share}%")
```

```python
#Удалим неявные дубликаты
df_all = df_all.drop_duplicates(subset=df_all.columns.drop('order_id'), keep='first').reset_index(drop=True)
```

```python
#Проведем преобразование данных в формат datetime
for column in ['created_dt_msk','created_ts_msk', 'data']:
    df_all[column] = pd.to_datetime(df_all[column])
```

```python
#Приведем выручку с заказа к единой валюте — российскому рублю.
for index, row in df_all.iterrows():
    if row['currency_code'] == 'kzt':
        df_all.at[index, 'revenue_rub'] = df_all.at[index, 'revenue']/df_all.at[index, 'curs']
    elif row['currency_code'] == 'rub':
        df_all.at[index, 'revenue_rub'] = df_all.at[index, 'revenue']   
```

```python
#Добавим столбец, где выручку с продажи одного билета на мероприятие 
df_all['one_ticket_revenue_rub']=df_all['revenue_rub']/df_all['tickets_count']
```

```python
#Добавим столбец, содержащий месяц заказа
df_all['month']=df_all['created_dt_msk'].dt.month
```

```python
#Добавим столбец с информацией о сезонности
def create_season(x):
    if x == 12 or x == 1 or x == 2:
        return 'зима'
    elif x >= 3 and x <= 5:
        return 'весна'
    elif x >= 6 and x <= 8:
        return 'лето'
    else:
        return 'осень'
df_all['season']=df_all['month'].apply(create_season)
```

```python
count_deleted_rows= df.shape[0]- df_all.shape[0]
print(f"При предобработке данных удалено строк {count_deleted_rows}")
```

- Удалены строки, с пропусками данных в столбцах `event_type_description`, `event_type_main`, выбросы по столбцу `revenue` (отбор значений по 99 процентилю), дубликаты по бронированию билета без учёта идентификаторов заказа 30
- Проведено преобразования типов данных в столбцах с датой и временем
- Созданы столбцы:
* `revenue_rub` — выручка с заказа в рублях
* `one_ticket_revenue_rub` — выручка с продажи одного билета на мероприятие;
* `month` — месяц оформления заказа;
* `season` — сезонность.

 ### 3.1. Анализ распределения заказов по сегментам и их сезонные изменения

Для каждого месяца найдем количество заказов и визуализируем результаты.

Фиксируется увеличение заказов от июня к ноябрю 2024 года.

Для осеннего и летнего периодов сравним распределение заказов билетов по разным категориям: тип мероприятия, тип устройства, категория мероприятия по возрастному рейтингу. 

```python
df_summer=df_all.loc[df_all['season']=='лето']
df_fall= df_all.loc[df_all['season']=='осень']

events_summer=df_summer.groupby('event_type_main')['order_id'].count()/df_summer.shape[0] 
events_fall= df_fall.groupby('event_type_main')['order_id'].count()/df_fall.shape[0]
```

```python
events_compare = pd.concat([events_summer, events_fall], axis=1)
events_compare.columns = ['Лето', 'Осень']
events_compare = events_compare.fillna(0)  # если какие-то категории встречались только в одном сезоне

# Строим сравнительный bar chart
events_compare.plot(kind='bar', figsize=(10,6))
plt.title('Распределение заказов по типу мероприятия (доли)')
plt.ylabel('Доля заказов')
plt.xlabel('Тип мероприятия')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Сезон')
plt.show()
```

```python
device_summer=df_summer.groupby('device_type_canonical')['order_id'].count()
device_fall=df_fall.groupby('device_type_canonical')['order_id'].count()

# Создаем фигуру с двумя осями для двух диаграмм
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Строим круговую диаграмму для летнего сезона
ax1.pie(device_summer, labels=device_summer.index, startangle=90, autopct='%1.1f%%')
ax1.set_title('Лето')

# Строим круговую диаграмму для осеннего сезона
ax2.pie(device_fall, labels=device_fall.index, startangle=90,autopct='%1.1f%%')
ax2.set_title('Осень')

# Отображаем диаграммы
fig.suptitle('Распределение заказов по типам устройств и сезонам')
plt.show()
 
```

```python
age_summer=df_summer.groupby('age_limit')['order_id'].count()/df_summer.shape[0] 
age_fall= df_fall.groupby('age_limit')['order_id'].count()/df_fall.shape[0]

age_compare = pd.concat([age_summer, age_fall], axis=1)
age_compare.columns = ['Лето', 'Осень']
age = age_compare.fillna(0)  # если какие-то категории встречались только в одном сезоне

# Строим сравнительный bar chart
age_compare.plot(kind='bar', figsize=(10,6))
plt.title('Распределение заказов по возрастному ограничению и сезонам (доли)')
plt.ylabel('Доля заказов')
plt.xlabel('Возрастное ограничение')
plt.xticks(rotation=0)
plt.legend(title='Сезон')
plt.show()
```

```python
counted_mean_ticket = df_all.groupby(['event_type_main','season'])['one_ticket_revenue_rub'].mean().unstack()

# Строим сравнительный bar chart
counted_mean_ticket.plot(kind='bar', figsize=(10, 6))
plt.title('Средняя выручка с одного билета по типам мероприятий и сезонам')
plt.ylabel('Средняя выручка, руб.')
plt.xlabel('Тип мероприятия')
plt.xticks(rotation=0)
plt.legend(title='Сезон')
plt.show()
```

При сопоставлении заказов по сезонам:
- распределение заказов по типу устройств не различаются
- летом популярнее концерты, осенью - спортивные мероприятия и театр
- летом популярнее мероприятия 18+, а осенью 0+
- выручка с продажи одного билета в зависимости от типа мероприятия летом и осенью примерно одинаковые

### 3.2. Осенняя активность пользователей 

```python
pivot= df_fall.groupby('created_dt_msk').agg({
    'order_id':'count',
    'user_id':'nunique',
    'one_ticket_revenue_rub':'mean'
})
pivot['orders_per_user']=pivot['order_id']/pivot['user_id']
pivot = pivot.rename(columns={'order_id': 'orders',
                             'user_id': 'DAU'})
 
```

```python
pivot[['orders','DAU']].plot(kind='line', figsize=(10, 4))
plt.title('Динамика заказов и количества уникальных пользователей в осенние месяца')
plt.xlabel('Дата')
plt.ylabel('Значение')

plt.grid(True)
plt.show()
```

```python
pivot['one_ticket_revenue_rub'].plot(kind='line', figsize=(10, 4))
plt.title('Динамика средней стоимости одного билета в осенние месяца')
plt.xlabel('Дата')
plt.ylabel('Значение')

plt.show()
```

```python
pivot['orders_per_user'].plot(kind='line', figsize=(10, 4))
plt.title('Динамика заказов на человека')
plt.xlabel('Дата')
plt.ylabel('Значение')


plt.show()
```

В течение осенних месяцев DAU растет постепенно. 3 сентября, 31 сентября, 31 октября было сделано необычно много дешевых заказов,  пользователи делали в эти дни больше заказов, чем обычно (больше 6 заказов на человека 3.09 и 31.09, больше 4 заказов 31.09)

```python
weekday_activity=df_fall.groupby(df_fall['created_dt_msk'].dt.day_name())['user_id'].nunique()

# Сортируем дни недели по порядку. Указываем порядок дней недели
weekday_activity.index = pd.Categorical(weekday_activity.index, categories=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ordered=True)

# Сортируем по индексу
weekday_activity = weekday_activity.sort_index()

weekday_activity.plot(kind='line', figsize=(10, 4))
plt.title('Активность пользователей по дням недели')
plt.xlabel('Дата')
plt.ylabel('Значение')

weekday_activity.plot(kind='line',
                     figsize=(10,4))
plt.title('Активность пользователей по дням недели')
plt.ylabel('Кол-во пользователей')

plt.grid(False)
plt.show()
```

Пользователи активно покупают билеты со вторника по пятницу (более 4800 активных пользователей, совершивших заказы), в выходные дни и в понедельник активность падает.

## 4. Статистический анализ данных

Проверим две гипотезы, которые предполагают большую активность пользователей мобильных устройств


***Первая гипотеза***  звучит так: Среднее количество заказов на одного пользователя мобильного приложения выше по сравнению с пользователями стационарных устройств.

Нулевая гипотеза H₀: Среднее количество заказов пользователей мобильного приложения и стационарных устройств не различаются

Альтернативная гипотеза H₁: Средняя количество заказов пользователей мобильного приложения выше, и это различие статистически значимо.

```python
# Создаем два набора данных для примера
data1 = df_fall.loc[df_fall['device_type_canonical']=='mobile'].groupby('user_id')['order_id'].mean()
data2 = df_fall.loc[df_fall['device_type_canonical']=='desktop'].groupby('user_id')['order_id'].mean()
# Создаем фигуру и оси
fig, ax = plt.subplots(figsize=(10,5))

# Строим диаграмму размаха
ax.boxplot([data1, data2]) 

# Добавляем метки для каждой диаграммы
ax.set_xticklabels(['Mobile', 'Desktop'])

# Добавляем заголовок и подписи к осям
ax.set_title('Сравнение среднего кол-ва заказов на пользователя по типам устройств')
ax.set_ylabel('Значения')

# Отображаем диаграмму
plt.show()
```

Распределение статистики среднего кол-ва заказов на пользователя по типу устройства похожи. Дисперсии не были рассчитаны, но предположим, что они не равны. Поэтому будем использовать двухвыборочный t-test Уэлча. 

```python
alpha=0.05

results=st.ttest_ind(
    data1,
    data2,
    equal_var=False,
    alternative='greater'
)
print(f'p_value={round(results.pvalue,2)}')
if results.pvalue<alpha:
    print('Отвергаем нулевую гипотезу')
else: 
    print('Нет оснований отвергать нулевую гипотезу')
    
```

Следовательно  среднее количество заказов пользователей мобильного приложения и стационарных устройств статистически не различаются

<a class="anchor" id="7-bullet"></a>
***Вторая гипотеза***: Среднее время между заказами пользователей мобильных приложений выше по сравнению с пользователями стационарных устройств.   

Нулевая гипотеза H₀: Среднее время между заказами пользователей мобильного приложения и стационарных устройств не различаются

Альтернативная гипотеза H₁: Среднее время между заказами пользователей мобильного приложения выше, и это различие статистически значимо.

```python
df_fall_filtered = df_fall.loc[~pd.isna(df_fall['days_since_prev'])]
```

```python

data3 = df_fall_filtered.loc[df_fall_filtered['device_type_canonical']=='mobile'].groupby('user_id')['days_since_prev'].mean()
data4 = df_fall_filtered.loc[df_fall_filtered['device_type_canonical']=='desktop'].groupby('user_id')['days_since_prev'].mean()
# Создаем фигуру и оси
fig, ax = plt.subplots(figsize=(10,5))

# Строим диаграмму размаха
ax.boxplot([data3, data4]) 

# Добавляем метки для каждой диаграммы
ax.set_xticklabels(['Mobile', 'Desktop'])

# Добавляем заголовок и подписи к осям
ax.set_title('Сравнение распределения среднего времени между заказами пользователей по типам устройств')
ax.set_ylabel('Значения')

# Отображаем диаграмму
plt.show()
```

Применим тест Манна-Уитни, поскольку выборочные средние не имеют нормального распределения, есть хвосты

```python
from scipy.stats import ttest_ind, mannwhitneyu


# Выборка целевой метрики для группы A

metric3 = df_fall_filtered[df_fall_filtered.device_type_canonical=='mobile'].groupby('user_id')['days_since_prev'].mean()
metric4= df_fall_filtered[df_fall_filtered.device_type_canonical=='desktop'].groupby('user_id')['days_since_prev'].mean()
# Создаем фигуру и оси

# На каком уровне значимости проверяем гипотезу о равенстве распределений
alpha = 0.05 

# Применяем тест Манна — Уитни
stat_mw, p_value_mw = mannwhitneyu(
    metric3, 
    metric4, 
    alternative='greater' 
) 

if p_value_mw > alpha:
    print(f'pvalue={p_value_mw} > {alpha}')
    print('Нулевая гипотеза находит подтверждение!')
else:
    print(f'pvalue={p_value_mw} < {alpha}')
    print('Нулевая гипотеза не находит подтверждения!')
```

Следовательно принимаем альтернативную гипотезу: среднее время между заказами пользователей мобильного приложения выше, и это различие статистически значимо.
